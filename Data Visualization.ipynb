# Scatterplot

```python
#you can check the relation between input and ouput likE UNIT AND PLACE 
sb.scatterplot(x="UNIT",y="PLACE",data=df1)
plt.figure(figsize=(150,70))
plt.show()
```

```python
sb.scatterplot(x="UNIT",y="PLACE",data=df1)
plt.figure(figsize=(150,70))
plt.show()
```

# DISTPLOT MEANS DISTRIBUTION PLOT ITS PART OF HISTOGRAM

```python
sb.distplot(df1["UNIT"])
plt.show()
#histogram best for numericL Data
```

```python
#only show curve using hist method
sb.distplot(df1["UNIT"],hist=False)
plt.show()
```

```python
#using color atributes
sb.distplot(df1["UNIT"],color="red")
plt.show()
```

```python
#using kde for removing curve line
#using bins means its show all big values like following chart
sb.distplot(df1["UNIT"],color="red",bins=5,kde=False) 
plt.show()
```

```python
#Now create a plot verticaly 
sb.distplot(df1["UNIT"],color="red",vertical=True)
plt.show()
#verticaly means replace the x and y 
```

# JointPLOt

```python
sb.jointplot(x="UNIT",y="PLACE",data=df1)
plt.figure(figsize=(150,70))
plt.show()
#joint plot have both plot histplot and scatterplot attributes 
```

```python
sb.jointplot(x="UNIT",y="PLACE",data=df1,color="orange")#for change color
plt.figure(figsize=(150,70))
plt.show()
```

# BOXPLOT

```python

sb.boxplot(x="UNIT",y="PLACE",data=df1)#check the relationshio between UNIT AND PLACE
plt.figure(figsize=(150,70))
plt.show()
```

```python
#boxplot have different palettte like set1 etc.palatte put the same colour on ramdom three city
sb.boxplot(x="UNIT",y="PLACE",data=df1,palette="Set1")
plt.figure(figsize=(150,70))
plt.show()
```

```python
sb.boxplot(x="UNIT",y="PLACE",data=df1,color="green")#using color attributes fro all convert to green
plt.figure(figsize=(150,70))
plt.show() 
```

```python
#change the thickness using linewidth method
sb.boxplot(x="UNIT",y="PLACE",data=df1,color="green",linewidth=3)
plt.figure(figsize=(150,70))
plt.show() 
#now its diffrence between upper and lower image
```

```python
sb.boxplot(x="UNIT",y="PLACE",data=df1,hue="DATE") #check relation with the date column which date done.
plt.figure(figsize=(150,70))
plt.show() 
```
